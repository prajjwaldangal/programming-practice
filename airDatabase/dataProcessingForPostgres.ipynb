```python
def listUniqueValsForAttrs(attrNo):
     uniqueAttributes.clear()
     for line in lines[1:]:
             tmp = line.split(",")
             uniqueAttributes.add(tmp[attrNo-1])
```

```python
class bcolors:
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKCYAN = '\033[96m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
```

```python
attr_list = list(lines[0].split(","))
print(attr_list)
```

```python
l = len(lines[0].split(","))
for i in range(1, l+1):
    print(f"{bcolors.HEADER} attr: {attr_list[i-1]}", end=" ")
    listUniqueValsForAttrs(i)
    ls = [len(el) for el in uniqueAttributes]
    m = max(ls)
    print(f"attr_max_len: {m}{bcolors.ENDC}")
    print("list: ", uniqueAttributes)
```
